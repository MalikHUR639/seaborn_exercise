# Box plots

```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
mpg=sns.load_dataset('mpg')
mpg.head()
```

```python
sns.boxplot(x=mpg.origin,y=mpg.mpg,order=['japan','europe','usa'],hue=mpg.cylinders)
```

```python
sns.load_dataset('titanic')
```

```python
titanic=sns.load_dataset('titanic')
```

```python
sns.scatterplot(x= titanic.age ,y= titanic.fare ,data=titanic)
```

```python
sns.scatterplot(x= titanic.age ,y= titanic.fare ,data=titanic,hue=titanic.sex)
```

```python
plt.figure(figsize=(16,9))
sns.scatterplot(x= titanic.age ,y= titanic.fare ,data=titanic,hue=titanic.sex,size=titanic.who,style=titanic.who,sizes=(100,200)
               )
```

```python
plt.figure(figsize=(16,9))
sns.scatterplot(x= titanic.who ,y= titanic.fare ,data=titanic,hue=titanic.alive,size=titanic.who,style=titanic.alive,sizes=(100,200),
               palette='flare',alpha=0.7)
```
