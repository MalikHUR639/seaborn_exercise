```python
df = df.drop(columns=["Date","Events","Wspeed"],axis=1).set_index("city")
df
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(df)
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(df,annot=True,vmin=24,vmax=50)
```

# Pair Plot

```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
breast_dataset= load_breast_cancer()
breast_dataset
```

```python
cancer_d=pd.DataFrame(np.c_[breast_dataset['data'],breast_dataset['target']],
                     columns=np.append(breast_dataset['feature_names'],['target']))
cancer_d
```

```python
cancer_d=cancer_d.head(6)
```

```python
sns.pairplot(cancer_d)
```

```python
sns.pairplot(cancer_d,vars=['mean radius', 'mean texture', 'mean perimeter', 'mean area',
        'mean smoothness'],hue='target')
```
